```python
import numpy as np
import matplotlib.pyplot as plt
```

Boundary Temperature Conditions

```python
T_grid = 25     #Initial Grid temperature (Room Temperature)

T_Top = 100     #Temperature on Top of the plate
T_Right = 0     #Temperature on Right side of the plate
T_Bottom = 0    #Temperature on Left side of the plate
T_Left = 0      #Temperature on Bottom of the plate
```

Plate Dimensions

```python
X = 100         #Width of the Plate
Y = 200         #Height of the Plate
```

Grid Temperatures

```python
Grid=np.empty((X, Y))
Grid.fill(T_grid)

#Assigning Boundary Values 

Grid[0,:] = T_Top
Grid[:,(Y-1)] = T_Right
Grid[(X-1),:] = T_Bottom
Grid[:,0] = T_Left

print("Temperatures on Top of the plate : ", Grid[0,:])
print("Temperatures on Right side of the plate  : ", Grid[:,(Y-1)])
print("Temperatures on Bottom of the plate  : ", Grid[(X-1),:])
print("Temperatures on Left side of the plate : ", Grid[:,0])
```

```python
def heatPlot(ar):
  colorinterpolation = 100
  colourMap = plt.cm.jet

  plt.title("2D Plate - Contour Temperature")
  plt.contourf(ar,colorinterpolation,cmap=colourMap)
  plt.colorbar()
  plt.show()
```

```python
n = 1000        #Number of Iterations
delX = 1       #Grid Increment along X-Axis
delY = 1       #Grid Increment along Y-Axis

print("Initial Temperature of Grid")
heatPlot(Grid)

for iteration in range(n):
    for i in range(1,X-1,delX):
        for j in range(1,Y-1,delY):
            Grid[i,j]=0.25*((Grid[i+1][j])+(Grid[i-1][j])+(Grid[i][j+1])+(Grid[i][j-1]))
    if((iteration+1)%100 == 0):
      print("After Iteration  : ", iteration+1)
      heatPlot(Grid)
    else:
      continue
```
